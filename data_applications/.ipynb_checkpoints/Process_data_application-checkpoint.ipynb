```python
import pandas as pd
import matplotlib.pyplot as plt
import os
import numpy as np

# Folder containing CSV files
folder_path = "results_data_application_seperate_quantile_OLS"

# List CSV files
csv_files = [f for f in os.listdir(folder_path) if f.endswith(".csv")]

# Initialize dictionaries to store data
data_linear = {}
data_quantile = {}

# Read CSVs
for file in csv_files:
    race = file.replace("coef_", "").replace(".csv", "")
    df = pd.read_csv(os.path.join(folder_path, file))
    df.set_index("Variable", inplace=True)
    data_linear[race] = df["Linear"]
    data_quantile[race] = df["Quantile"]

# Variables of interest
variables = ["prenatal_visit_category6-10", "prenatal_visit_category>10"]
races = [r for r in data_linear.keys() if r != "white"]

# Compute differences with White
diff_linear = pd.DataFrame({
    race: data_linear[race].loc[variables] - data_linear["white"].loc[variables] 
    for race in races
})
diff_quantile = pd.DataFrame({
    race: data_quantile[race].loc[variables] - data_quantile["white"].loc[variables] 
    for race in races
})

# Optional: rename index for plotting
diff_linear.index = ["6-10", ">10"]
diff_quantile.index = ["6-10", ">10"]

# Plot function mimicking the attached style
def plot_grouped_bars(diff_df, title, filename = None):
    bar_width = 0.25
    x = np.arange(len(diff_df.index))
    
    fig, ax = plt.subplots(figsize=(8,5))
    
    colors = ["#1f77b4", "#ff7f0e", "#2ca02c"]  # Blue, Orange, Green
    
    for i, race in enumerate(diff_df.columns):
        ax.bar(x + i*bar_width, diff_df[race], width=bar_width, label=race, color=colors[i])
    
    ax.set_xticks(x + bar_width / 2)
    ax.set_xticklabels(diff_df.index)
    ax.set_xlabel("Number of prenatal visits")
    ax.set_ylabel(r"$\hat{\beta}_1 - \hat{\beta}_0$")
    ax.set_title(title)
    ax.axhline(0, color='black', linewidth=0.8)
    ax.legend(frameon=True, loc='lower left')
    plt.tight_layout()
    if filename: 
        plt.savefig(filename, format='pdf')
    plt.show()

# Plot OLS differences
plot_grouped_bars(diff_linear, "OLS: Difference from White", "prenatal_visit_difference_OLS.pdf")

# Plot Quantile differences
plot_grouped_bars(diff_quantile, "Quantile Regression: Difference from White","prenatal_visit_difference_quantile.pdf")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import os
import numpy as np

# -------------------------
# 1. OLS and Quantile differences from CSVs
# -------------------------
folder_path = "results/results_data_application_seperate_quantile_OLS"
csv_files = [f for f in os.listdir(folder_path) if f.endswith(".csv")]

data_linear = {}
data_quantile = {}

for file in csv_files:
    race = file.replace("coef_", "").replace(".csv", "")
    df = pd.read_csv(os.path.join(folder_path, file))
    df.set_index("Variable", inplace=True)
    data_linear[race] = df["Linear"]
    data_quantile[race] = df["Quantile"]

variables = ["prenatal_visit_category6-10", "prenatal_visit_category>10"]
races = [r for r in data_linear.keys() if r != "white"]

diff_linear = pd.DataFrame({
    race.capitalize(): data_linear[race].loc[variables] - data_linear["white"].loc[variables]
    for race in races
})

diff_quantile = pd.DataFrame({
    race.capitalize(): data_quantile[race].loc[variables] - data_quantile["white"].loc[variables]
    for race in races
})

x_labels = ["6-10", ">10"]

# -------------------------
# 2. Disparities from separate CSV
# -------------------------
data = pd.read_csv("results/results_data_application_seperate2.csv")
data_tau = data[data["tau"]==0.05]
data_05 = data_tau[["race","prenatal_visit_category6.10","prenatal_visit_category.10"]]

disparities = {}
for race in ["black", "asian", "hispanic"]:
    tmp = data_05[data_05["race"]==race].drop(columns=["race"]).reset_index(drop=True)
    if not tmp.empty:
        disparities[race.capitalize()] = tmp.iloc[0] - data_05[data_05["race"]=="white"].drop(columns=["race"]).reset_index(drop=True).iloc[0]

df_disparity = pd.DataFrame(disparities)

# -------------------------
# 3. Plot all three side by side without 'index' column
# -------------------------
colors = ["#1f77b4", "#ff7f0e", "#2ca02c"]
bar_width = 0.25

fig, axes = plt.subplots(1, 3, figsize=(20, 6), sharey=True)

plot_data = [
    (diff_linear, "OLS"),
    (diff_quantile, "Quantile Regression"),
    (df_disparity, "ES Regression")
]

for ax, (df, title) in zip(axes, plot_data):
    x = np.arange(len(x_labels))
    for i, col in enumerate(df.columns):
        ax.bar(x + i*bar_width, df[col], width=bar_width, label=col, color=colors[i % len(colors)])
    ax.set_xticks(x + bar_width)  # center bars
    ax.set_xticklabels(x_labels, rotation=0, fontsize=14)
    ax.set_title(title, fontsize=16)
    ax.axhline(0, color='black', linewidth=0.8)
    ax.set_xlabel("Number of prenatal visits", fontsize=14)
    ax.legend(loc='lower left', fontsize=12)

axes[0].set_ylabel(r'$\hat \beta_1 - \hat \beta_0$', fontsize=16, rotation=0, labelpad=50)

plt.tight_layout()
plt.savefig("prenatal_visit_difference_three_plots.pdf", format='pdf')
plt.show()
```

```python
import pandas as pd
data = pd.read_csv("data.csv")
data = data[data["race_mom"].isin([1,2,4,7])]
```

```python
import numpy as np
import matplotlib.pyplot as plt

qs = [0.01, 0.1, 0.5]
quantile_curves = {q: [] for q in qs}
means = []
order = range(15)

for cat in order:
    vals = data.loc[data["prenatal_visit"] == cat, "birthweight"].dropna().values
    if len(vals) == 0:
        for q in qs: 
            quantile_curves[q].append(np.nan)
        means.append(np.nan)
    else:
        for q in qs:
            quantile_curves[q].append(np.quantile(vals, q))
        means.append(np.mean(vals))

plt.figure(figsize=(5,4))

linestyles = ['-', '--', ':']  # Different styles for each quantile
markers = ['D', 'o', '^']      # (Optional) different markers

# Plot quantile curves with different linestyles
for i, (q, vals) in enumerate(quantile_curves.items()):
    plt.plot(order, vals, 
             marker=markers[i % len(markers)], 
             linestyle=linestyles[i % len(linestyles)],
             label=f"{q:.1} quantile",
             linewidth=2)

plt.xlabel("Number of prenatal visits", fontsize=14)
plt.ylabel("Birthweight (g)", fontsize=14)
plt.tight_layout()
plt.savefig("birthweight_by_prenatal_visit.pdf", dpi=300)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Desired order
order = ["African American", "Asian", "Hispanic", "White"]
pos = {cat: i for i, cat in enumerate(order)}

# Keep only rows we care about
df = data[data["race_mom"].isin(order)].copy()
df["x"] = df["race_mom"].map(pos).astype(float)   # ensure numeric

plt.figure(figsize=(9,6))

# Optional jitter so points spread out horizontally
rng = np.random.default_rng(0)
jitter = rng.uniform(-0.15, 0.15, size=len(df))

# Scatter raw data
plt.scatter(df["x"] + jitter, df["birthweight"], alpha=0.4, s=14, label="Data")

# Quantiles (0.1%, 1%, 5%)
qs = [0.001, 0.01, 0.05]
for cat, x in pos.items():
    vals = df.loc[df["race_mom"] == cat, "birthweight"].dropna().values
    if len(vals) == 0:
        continue
    qv = np.quantile(vals, qs)
    plt.scatter([x]*len(qs), qv, marker="D", s=70,
                edgecolor="k", linewidth=0.5, color="red", label="_nolegend_")
    
    # Mean
    mean_val = np.mean(vals)
    plt.scatter(x, mean_val, marker="s", s=90,
                edgecolor="k", linewidth=0.5, color="blue", label="_nolegend_")

# Format axes
plt.xticks(list(pos.values()), order)
plt.xlim(-0.5, len(order)-0.5)
plt.xlabel("Mother's Race")
plt.ylabel("Birthweight")
plt.title("Birthweight by Mother's Race with Tail Quantiles & Mean")

# Legend
from matplotlib.lines import Line2D
plt.legend(handles=[
    Line2D([0],[0], marker="o", linestyle="None", label="Data", color="gray"),
    Line2D([0],[0], marker="D", linestyle="None", label="Quantiles (0.1%, 1%, 5%)", color="red"),
    Line2D([0],[0], marker="s", linestyle="None", label="Mean", color="blue")
], loc="best")

plt.tight_layout()
plt.show()
```
